```python
#(2) plot the number of AirBnbs per neighborhood
hood_list = []
homes_list = []
for i in df["neighbourhood"].unique():
    num_of_homes1 = int(len(i)) #just making sure that the sum is actually an integer lol
    hood_list.append(i)
    homes_list.append(num_of_homes1)
    #print(i, num_of_homes1) #this prints the neighborhood and number of homes!
    
fig = plt.figure(figsize = (10,5))
plt.bar(hood_list, homes_list, color='pink', width=0.4)
plt.xlabel(("Neighborhoods in Amsterdam"))
plt.ylabel("Number of AirBnBs")
plt.title("Number of AirBnBs per Neighborhood in Amsterdam")
print("Answer 2: The number of AirBnBs per neighborhood is listed below:")
plt.xticks(rotation=90)
plt.show()

#how to rotate the x labels?

    
```
